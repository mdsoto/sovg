* Text (columnar data separated by space, commas...)

* Image files (bmp, jpg, tiff,...)

As is common in Python, there are different ways to input or output information, depending on the format of the files and the library or module you use. In this and the following session, we are going to explore simple ways to get information (text files and images) in and out of Python.

## Reading text files

Text files are the most straightforward files used in any computer (regardless of the operative system) to save information in a format familiar to us. Some specific uses of the text files are:

* Simple documents like notes, letters, emails, ...
* Data in columnar shape
* Keepers of parameters or environment variables for programs or operative systems
* Scripts or programs on different programming languages
* Communication between machines, programs, and persons
* Databases

Characters in text files were initially coded for electronic communication (teleprinters) based on different standards that in 1963 became the **ASCII** (American Standard Code for Information Interchange) code. Soon, it was clear that the ASCII code, which only had English characters, some mathematical,  and special symbols (128 characters), was insufficient for other written languages. So the ASCII code started to evolve into different codes until we got to the actual **UTF-8** (Universal Transformation Format) which has 1,112,064 characters, enough for all human written languages and disciplines. The UTF-8 standard is the base on which the World Wide Web works.

More information on character encoding at: https://www.w3schools.com/charsets/

### Reading multi-column text files with missing values

To have a file with just a column is very rare. Normally we have files with several numerical columns, sometimes even columns with text. When you load a multi-column file an array is created with the corresponding number of columns, later each one can be defined as independent variables. Important, to operate with these variables, they have to have the same dimensions (same number of rows).

A common situation is to have several missing or lost values, places where the measurements were not possible. These missing values used to be indicated with special numbers such as -999.000, or -999.2500, or just with empty spaces. Numpy recognizes automatically empty spaces as missing values and replaces them with **nan**, not a number.

We are going to load the *rampi.csv* file, also in the Input folder, which is a five-column file that comes from a study of gold and associated elements (in ppm) in quartz veins in the **Rampi block prospect**, Indonesia (see image below from Google Earth):

### Writing single or multi-column text files

After manipulating your data you could have the necessity of exporting your results to other users or programs. The previous file is more suitable for single or few variables. For massive variables with a lot of rows, it is more convenient to export them in single or multi-column text files. Contrary to the previous examples, here it is important to take care of the format in which those variables are written. Maybe you need just four decimals or scientific notation. Here are two references about numbers and strings format:

https://pyformat.info/

https://mkaz.blog/code/python-string-format-cookbook/

```python
# For calculating ratios the columns have to have the same length

cuau_rat = cu/au
zn2au_rat = zn2/au
pbau_rat = pb/au
# znau_rat = zn/au

cuau_rat
```

```python
# Writing a ratio to a single-column file

np.savetxt('Output/cuau_rat.txt', cuau_rat, header='cuau ratio',fmt='%1.4f')

# fmt='%1.4f' write a float with this format: 20.2431
```

### Bar and pie plots

Here you have two more types of plots, bar and pie plots. The categories (elements), data (sizes), and colors are the same in both cases. 

```python
# Bar plot

# Data to plot

elements = ['Au', 'Ag', 'Cu', 'Zn', 'Pb']
sizes = [np.nanmean(au), np.nanmean(ag), np.nanmean(cu), np.nanmean(zn2), np.nanmean(pb)]
x = range(len(elements))

# Plot properties
    
colors = ['y', 'c', 'r','m','b']

########################### Do not change the code below #####################################

# Plot & identifiers
    
plt.figure(figsize=(8,5))

plt.bar(x, sizes, align='center', color=colors, alpha=0.5)

plt.xticks(x, elements)

# Labels

plt.title('Rampi prospect')
plt.xlabel('Elements')
plt.ylabel('Average concentration (ppm)')
plt.show()
```

## Multiplots by subplot

With the Matploplib's **subplot** function it is possible to gather several plots in a composition, arranged based on a rectangular grid of raw and columns. The arguments (2,2,1) in the following command:

`plt.subplot(2,2,1)`
    
means the first plot in an arrange of two rows and two columns of plots, something like this:

<img src="NB_images\multiplot_example.png"  style="width:700px">

```python
# Building and saving a 2 x 3 multiplot with the Rampi data

plt.figure(figsize=(16,10))

plt.subplot(2,3,1)

# Point-line plot
plt.plot(au, 'y', label="Au")
plt.plot(ag, 'c', label="Ag")
plt.plot(cu, 'r', label="Cu")
plt.plot(zn2,'m', label="Zn")
plt.plot(pb, 'b', label="Pb")
plt.title("Elements")
plt.xlabel("File order")
plt.ylabel("ppm")
plt.legend()
plt.grid(True)

plt.subplot(2,3,2)

# Bar plot
elements = ('Au', 'Ag', 'Cu', 'Zn', 'Pb')
sizes = [np.nanmean(au), np.nanmean(ag), np.nanmean(cu), np.nanmean(zn2), np.nanmean(pb)]
x = np.arange(len(elements))   
colors = ['y', 'c', 'r','m','b']
plt.bar(x, sizes, align='center', alpha=0.5, color=colors)
plt.xticks(x, elements)
plt.title('Elements')
plt.ylabel('ppm')
plt.xlabel('Elements')
plt.grid(True)

plt.subplot(2,3,3)

# Cross plot
plt.loglog(au,ag,'o')
plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)

plt.subplot(2,3,4)

# Cross plot
plt.loglog(au,cu,'o')
plt.title("Au vs. Cu")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Cu (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,5)

# Cross plot
plt.loglog(au,zn2,'o')
plt.title("Au vs. Zn")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Zn (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,6)

# Cross plot
plt.loglog(au,pb,'o')
plt.title("Au vs. Pb")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Pb (ppm)")
plt.ylim(1,200)
plt.grid(True)

# Avoid the overlap between labels

plt.tight_layout()

# Saving multiplot to file

plt.savefig("Output/multiplot.png")
plt.savefig("Output/multiplot.pdf")

plt.show()
```

## Multiplot by looping an array

Multiplots can be built by a loop, making the code shorter and more efficient. Let's see an example done by slicing the Rampi array: 

```python
rampi.shape
```

```python
# This in gold

rampi[:,0]
```

```python
# Elements, colors, and position in the vein, according to the rampi array

elements = ['Au', 'Ag', 'Cu', 'Zn', 'Pb']
colors = ['y', 'c', 'r', 'm', 'b']
x = range(0, 3*rampi.shape[0], 3)

# Loop according to the number of columns in the rampi array

for i in range(rampi.shape[1]):
    
    plt.figure(figsize=(16,4))
    
    # Line plot of the element
    plt.subplot(1,2,1)
    plt.plot(x, rampi[:,i], color=colors[i], alpha=0.5)
    plt.xlim(0, 65)
    plt.xlabel('Position in the vein (m)')
    plt.ylabel('Concentration (ppm)')
    plt.title('Plot of '+elements[i])
    plt.grid()
    
    # Histogram of the element
    plt.subplot(1,2,2)
    plt.hist(rampi[:,i], color=colors[i], alpha=0.5)
    plt.ylabel('Frequency')
    plt.xlabel('Concentration (ppm)')
    plt.title('Histogram of '+elements[i])
    plt.grid()
    plt.savefig("Output/plot_"+elements[i]+".png")  
     
plt.tight_layout()
plt.show()        
```

## Widgets

Widgets are interactive controls or GUI tools that bring your Jupyter Notebook to life. There are many sources and types of widgets, however, all of them are focused on facilitating the interaction between the user and the notebook, minimizing the direct modification of the code. The main widgets are:

* Tabs, accordions, horizontal and vertical layout boxes, grid layouts
* Advanced controls like maps, 2d and 3d visualizations, data grids, and more

One of the main libraries of widgets is: https://ipywidgets.readthedocs.io/en/stable/

Let's see an example of widgets, an interactive slider, but first we need to install the necessary libraries:

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from matplotlib.widgets import Cursor
```

```python
# Sine plot with an interactive slider

# Plotting function

def plot_func(freq):
    plt.figure(figsize=(11,5))
    x = np.arange(0, 2*np.pi, 0.01) # 2*np.pi is 360º
    y = np.sin(x * freq)
    plt.plot(x, y)
    plt.xlabel('Radians')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.show()

# The widget receives the plotting function

slider = widgets.FloatSlider(value=1, min=1, max=20, step=0.1, description="Frequency")
interact_plot = widgets.interactive(plot_func, freq=slider)

# Display the widget
display(interact_plot)
```

## Procedural vs. object-oriented programming

Before covering the last topic is worth seeing the difference between building a multiplot using the procedural and the object-oriented programming (OOP) paradigms. Here we are going to do the same scatter plot, with the same data, with both paradigms of coding:

```python
# Plot built with the procedural paradigm

plt.figure(figsize=(8,3))
plt.scatter(au, ag)    # The following three lines are equivalent to plt.loglog(au,ag,'.')

plt.yscale('log')
plt.xscale('log')

plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)
plt.show()
```

```python
# Plot built with the OOP paradigm

fig, ax = plt.subplots(figsize=(8, 3))
ax.scatter(au,ag)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_title("Au vs. Ag")
ax.set_xlabel("Log Au (ppm)")
ax.set_ylabel("Log Ag (ppm)")
ax.grid(True)

plt.show()
```

In code length both styles are equivalent. In the availability of examples and power, the **oop paradigm** (more modern) **is the winner** and is where we have to aim. A step up from the Jupyter Notebook widgets is Qt5, which is a cross-platform (Windows, Mac, Linux) UGI toolkit for application development (Spider for example). One of its many applications is that allows the plots to be independent from the notebook. More information on Qt5 at:

```python
# Plot built with the procedural paradigm

# The num=1 between the brackets identifies the new plot window
# It is needed to avoid overwriting of the window

plt.figure(num=1, figsize=(8,5)) 
plt.scatter(au, ag)
plt.yscale('log')
plt.xscale('log')
plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)
plt.show()
```
