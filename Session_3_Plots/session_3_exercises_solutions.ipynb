In the section called "The Mineralogy of Sedimentary Rocks" of the following website there is a table with the mineral composition of two types of sedimentary rocks (mudrocks and sandstones):

http://www.tulane.edu/~sanelson/eens212/sedrxintro.htm

Built a bar and a pie plot for both rocks considering the following:
    
* Make a self-contain data code which doesn't depend on external files
* Make the necessary assumptions to represent all the components 
* Use the right scales, title, labels, and grid

```python
# Data and plots parameters

minerals = ['Clays', 'Quartz', 'Feldspar', 'Carbonates', 'OrganicM & others']

Muds_Perc = [60, 30, 4, 3, 2.5]
Sands_Perc = [5, 65, 15, 5, 5]

x = np.arange(len(minerals))

plotColours = ['y', 'c', 'r', 'm', 'b']
```

```python
# Bar plot

plt.figure(figsize = (17, 6))
    
# Muds Plot & identifiers:
plt.subplot(121)
plt.bar(x, Muds_Perc, align = "center", color = plotColours, alpha = 0.5)
plt.xticks(x, minerals)
plt.title("Mud Rocks minerals")
plt.ylabel("%")

# Sst Plot & identifiers
plt.subplot(122)
plt.bar(x, Sands_Perc, align = "center", color = plotColours, alpha = 0.5)
plt.xticks(x, minerals)
plt.title("Sandstones minerals")
plt.ylabel('%')

plt.tight_layout()
plt.show()
```

```python
# Another compact alternative for bar

plt.figure(figsize=(10,6))

plt.bar(x, Muds_Perc, color = "y", width = 0.25, label='Mudrocks')
plt.bar(x + 0.25, Sands_Perc, color = "orange", width = 0.25, label='Sandstones')

plt.xticks(x + 0.125, minerals)

plt.legend()
plt.title('mineral_Composition')
plt.ylabel('%')
plt.show()
```

```python
# Pie plot:

plt.figure(figsize=(17,6))

# Muds Plot & identifiers:

explode = (0.1,0,0,0,0)

plt.subplot(121)
plt.pie(Muds_Perc, explode=explode, labels=minerals, colors=plotColours, wedgeprops={'alpha':0.5},
        autopct='%1.1f%%', shadow=False, startangle=0)
plt.axis('equal')
plt.title("Mud Rocks minerals")

# Sst Plot & identifiers:

explode = (0,0.1,0,0,0)

plt.subplot(122)
plt.pie(Sands_Perc, explode=explode, labels=minerals, colors=plotColours, wedgeprops={'alpha':0.5},
        autopct='%1.1f%%', shadow=False, startangle=0)
plt.axis('equal')
plt.title("Sandstones minerals")
plt.show()
```

Load the file as an array and build a for loop for plotting two plots for each variable (could be the same plots used in the multiplots by loop example or others). The last variable, Potability, is a categorical variable (only 0 for non-potable and 1 for potable). What can you infer from the plots of this variable regarding the file structure?

```python
# Data loading into an array

wat_pota = np.genfromtxt('Input/water_potability.csv', skip_header = 1, delimiter=',')
wat_pota
```

```python
# Rows and columns

wat_pota.shape
```

```python
# Basic plot by for loop

    # Variables names coming from the file

names = ['ph','Hardness','Solids','Chloramines','Sulfate','Conductivity','Organic_carbon','Trihalomethanes','Turbidity', 'Potability']

    # Loop according to the number of columns in the array

for i in range(wat_pota.shape[1]):
    
    plt.figure(figsize=(16,4))
    
    # Line plot of the element
    plt.subplot(1,2,1)
    plt.plot(wat_pota[:,i], '.')
    plt.xlabel('File position')
    plt.ylabel(names[i])
    plt.title('Plot of '+names[i])
    plt.grid(True)
    
    # Histogram of the element
    plt.subplot(1,2,2)
    plt.hist(wat_pota[:,i], bins=50)
    plt.ylabel('Frequency')
    plt.xlabel(names[i])
    plt.title('Histogram of '+names[i])
    plt.grid(True)
     
plt.show()        
```

The samples in the file are grouped in batches of Potability 0, 1, 0, 1, ... That is why you see out-of-phase segments of 0 and 1 in the left plot. In general, the variables present continuous ranges of values that do not allow differences between potable and non-potable water.

</div>

```python
# Plotting function of a simple polynomial

def poly_plot(a, b, c):
    plt.figure(figsize=(11,5))
    
    x = np.linspace(-10, 10, 100)
    
    y = a*x**b + c
    
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(0, 100)
    plt.xlim(-10, 10)
    plt.title('Polynomial a*x**b + c')
    plt.grid(True)
```

```python
# 3 slicers, for a, b and c

widgets.interact(poly_plot,
                 a = widgets.FloatSlider(value=1, min=1, max=50, step=1), 
                 b = widgets.FloatSlider(value=2, min=1, max=10, step=1),
                 c = widgets.FloatSlider(value=10, min=1, max=100, step=1))
plt.show()
```
