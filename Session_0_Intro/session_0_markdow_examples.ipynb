# Markdown

Markdown is a lightweight and easy-to-use markup language primarily used to format text in plain text files. It was created by John Gruber in 2004 to be simple and readable in its raw form and when converted to HTML.

Key Features:

* Simple syntax: Focuses on being easy to read and write without complex tags.
* HTML conversion: It can be easily transformed into HTML for web pages.
* Widely supported: Tools like GitHub, Notion, Reddit, and many blogs support it.

More information at: 

https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html  <br>
https://daringfireball.net/projects/markdown

## Mermaid Flow Charts

Mermaid flowcharts are a way to create diagrams (such as flowcharts, sequence diagrams, or mind maps) using plain text in Markdown cells, rather than drawing them manually. They’re part of Mermaid, a JavaScript-based tool that turns simple text descriptions into visual diagrams automatically.

More information at: 

https://mermaid.js.org/#/ <br>
https://mermaid.ai/open-source/syntax/flowchart.html

```mermaid
graph TD
A[Start] --> B{Decision}
B -->|Yes| C[Do something]
B -->|No| D[Do something else]
```

```mermaid
flowchart TB
    subgraph "<strong> Vertical Flow Chart</strong>"
        A[Original Data] -->|3rd Filter| B[Filter Data] -->|Anomaly Filter| C[Filter Data without Anomalie]
    end

    C -->|KNN Imputer| D[Imputed Filter Data]
```
</br>

```mermaid
flowchart TB

subgraph "<strong> Horizontal Flow Chart</strong>"

A(Legacy GR 008) --> C("<strong> Integrated Data 008</strong> Features + GR") -->|Train| D("<strong> f(x):</strong>  Regression Function")
B(<strong> Imputed Data 008</strong> Features) --> C

style B fill:#98FB98,stroke:#333,stroke-width:1px
style C fill:#2E8B57,stroke:#333,stroke-width:1px
end
```

# Tables

Markdown tables are a simple way to organize data into rows and columns using plain text, where vertical bars `|` separate columns and a row of dashes defines the header, making them easy to read and write. They support basic alignment using colons (left, center, right) and are ideal for documentation or quick data display, although they have limitations such as no support for merged cells or advanced styling depending on the renderer.

More information at: https://saturncloud.io/blog/tables-in-markdown-in-jupyter/
