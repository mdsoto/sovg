Between 2000 and 4000 km depth (d in km), the temperature (temp in °C) inside the Earth can be estimated by the following equation:
<br/>
<br/>   
$$temp = (-1.12x10^{-12})d^4 + (2.85x10^{-8})d^3 + (-0.00031)d^2 + 1.64d + 930$$
<br/>   
Using the scientific notation (1e3 = 1000, or 1e-3 = 0.001) for the coefficients of the equation (a, b, c, and e), compute the temperature at 2800 km with only three decimals.

<br/>    
Given the following system of equations (3 equations and 3 unknowns):
<br/>
<br/>
\begin{align}
2x + 4y + 3z = 8  \\
3x + 5y + 6z = 7  \\
x +  3y - 2z = 5  
\end{align}
<br/>  
Use linear algebra operations to solve the following system of equations. Remember that the system can be represented as a dot product of the corresponding arrays, with the matrix A being the coefficients of the equation, u the unknown coordinates, and B the vector with the results:

\begin{align}
A \cdot u = B
\end{align}


The solution will be:

\begin{align}
u = A^{-1} \cdot B
\end{align}

where $A^{-1}$ is the inverse matrix of A
</div>

```python
A = [[2, 4, 3], [3, 5, 6], [1, 3, -2]]
A = np.array(A)
print('A = ', A)
print('Shape of A: ', A.shape, '\n')

B = [8,7,5]
B = np.array(B)
print('B = ', B)
print('Shape of B: ', B.shape)
```

```python
inv_A = np.linalg.inv(A)
print('Inverse of A: ', '\n', inv_A)
print('Shape of inverse A: ', inv_A.shape,'\n')

u = np.dot(inv_A,B)
print('The solution u is:','\n', u)
print('Shape of inverse u: ', u.shape,'\n')

print('Verification, B is:','\n', np.dot(A,u))
#print('Is Au=B: ', np.dot(A,u) == b)
```
