<b>Exercise 11: </b>
<br/>
<br/>    
Given the following system of equations (3 equations and 3 unknowns):
<br/>
<br/>
\begin{align}
2x + 4y + 3z = 8  \\
3x + 5y + 6z = 7  \\
x +  3y - 2z = 5  
\end{align}
<br/>  
Use linear algebra operations to solve the following system of equations. Remember that the system can be represented as a dot product of the corresponding arrays, with the matrix A being the coefficients of the equation, u the unknown coordinates, and B the vector with the results:

\begin{align}
A \cdot u = B
\end{align}


The solution will be:

\begin{align}
u = A^{-1} \cdot B
\end{align}

where $A^{-1}$ is the inverse matrix of A
