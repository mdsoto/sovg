* **Create a numpy random array (1x20)**
* Multiply the array by 5
* Print between the sixth  to the tenth element, both included
* Print every two elements of the complete array
* Make an example of slicing using positive and negative steps
    
Note: an array can be a matrix (2D) or a vector (1D)
</div>

```python
# note that these options are different

print(np.random.rand(20).shape)
print(np.random.rand(20, 1).shape)
print(np.random.rand(1, 20).shape)
print(np.random.rand(20, 20).shape)
```

```python
# Random vector of 20 elements:

np.random.seed(1) # seed is for getting the same random numbers
vect = np.random.rand(20)
print("Original vector:\n", vect, '\n')

# Multiply by 5:

vect = vect*5
print("Vector by 5:\n", vect, '\n')

# 6th-10th elements, both included:

print("6th-10th elements:\n", vect[5:10], '\n')

# Every 2 elements:

print("Every 2\n", vect[::2], '\n')  # Another way: print(vect[::2])

# Slicing using positive and negative steps:

print("Positive steps\n", vect[1:21:1], '\n')
print("Negative steps\n", vect[21:0:-1])
```

<font size="6"> &#127947;&#127997;</font> <b>Exercise 10: </b> 
    
Based on the previous functions to convert between the two temperature scales, write a small program that allows the user the following:

* Enter the temperature scale of the input data
* Enter the temperature value
* Print the equivalent temperature in the other scale.

</div>

```python
def convert_temps():
    
    """Function to convert between the two temperature scales, C to F or F to C"""

    scale = input("Please enter temperature scale, f or c: \n")
    temp = float(input("Please enter temperature value: \n"))

    if scale == 'f' or scale == 'F':
        tempc = (5/9)*(temp - 32)
        print('Temperature in ºC is: ', tempc)
    else:
        tempf = (9/5)*temp + 32
        print('Temperature in ºF is: ', tempf)
```

```python
convert_temps()
```

Generate a list  with angles from 0 to 359 degrees with step 2 degrees. Using a for loop calculate a vector with the cosine values of the angles. Try to do a plot.


</div>

```python
# In order to get the cosine, angles must be in radians

angles = np.arange(0, 360, 2)
radians = angles*(np.pi/180)
cos = []

for rad in radians:
    cos.append(np.cos(rad))

plt.plot(angles, cos)
plt.xlabel('Angle')
plt.ylabel('Consine')
plt.show()
```

<font size="6"> &#127947;&#127997;</font> <b>Exercise 14: </b>Create a program that computes n terms of the Fibonacci series. Try to do a simple plot of the serie.
</div>

```python
# With a for loop (4 lines)

fibo = [1, 1]
terms = range(0, 26)

for i in range(2, len(terms)):
    fibo.append(fibo[-1] + fibo[-2]) # Adding the sum of the last two elements to the fibo list

print("Fibonacci's sequence:\n", fibo, '\n')
plt.plot(terms, fibo)
plt.xlabel('Index')
plt.ylabel('Fibonacci number')
plt.show()
```

```python
# With a while loop (6 lines)

fibo = [1, 1]
terms = range(0, 26)

i = 2
while i < len(terms):
    fibo.append(fibo[-1] + fibo[-2])
    i = i + 1

print("Fibinacci's sequence :\n", fibo, '\n')
plt.plot(terms, fibo)
plt.xlabel('Index')
plt.ylabel('Fibonacci number')
plt.show()
```
